```python
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")

day_counts = tips['day'].value_counts()

plt.figure(figsize=(7, 7))
plt.pie(
    day_counts.values,
    labels=day_counts.index,
    autopct='%1.1f%%',
    colors=sns.color_palette("pastel"),
    wedgeprops={'edgecolor': 'black'}
)
plt.title('Proportion of Transactions by Day of the Week')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = {
    'total_bill': [79.25, 8.32, 5.06],
    'tip': [8.32, 1.91, 0.64],
    'size': [5.06, 0.64, 0.90]
}

covariance_matrix = pd.DataFrame(data, index=['total_bill', 'tip', 'size'])

plt.figure(figsize=(8, 7))
sns.heatmap(
    covariance_matrix,
    annot=True,
    fmt=".2f",
    cmap='coolwarm',
    linewidths=.5,
    linecolor='black',
    cbar_kws={'label': 'Covariance Value'}
)
plt.title('Covariance Matrix Heatmap for Tips Data')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


survival_by_class = pd.crosstab(titanic['Pclass'], titanic['Survived'])


survival_by_class.plot(
    kind='bar',
    stacked=True,
    figsize=(8, 6),
    color=['lightcoral', 'mediumseagreen']
)
plt.title('Count of Survival Status within Each Passenger Class ')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

palette = {'Yes': 'mediumseagreen', 'No': 'lightcoral'}

fig, axes = plt.subplots(1, 2, figsize=(15, 6))
fig.suptitle('Survival Analysis on Titanic ', fontsize=16)


sns.barplot(
    x='Sex',
    y='Survived',
    data=titanic,
    hue='Sex',
    palette={'male': 'skyblue', 'female': 'hotpink'},
    errorbar=None,
    ax=axes[0]
)
axes[0].set_title('Mean Survival Rate by Gender')
axes[0].set_ylabel('Survival Rate (Mean of Survived)')
axes[0].set_xlabel('Gender')
axes[0].tick_params(axis='x', rotation=0)


sns.barplot(
    x='Pclass',
    y='Survived',
    data=titanic,
    palette='magma',
    errorbar=None,
    ax=axes[1]
)
axes[1].set_title('Mean Survival Rate by Passenger Class')
axes[1].set_ylabel('Survival Rate (Mean of Survived)')
axes[1].set_xlabel('Passenger Class (1=First, 2=Second, 3=Third)')
axes[1].tick_params(axis='x', rotation=0)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```
